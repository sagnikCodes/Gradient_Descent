```python
import numpy as np
```

```python
training_data=np.loadtxt("train.csv",delimiter=",")
input_data=training_data[:,:len(training_data[0])-1]
output_data=training_data[:,len(training_data[0])-1]
print(input_data.shape)
print(output_data.shape)
```

```python
#pehle list mei convert taaki can add 1 easily using append
input_data_list=input_data.tolist()
for i in range(len(training_data)):
    input_data_list[i].append(1)
#then wapis numpy array mei
input_data=np.array(input_data_list)
print(input_data.shape)
```

```python
#now print input and output and take it's length for later use
number_of_input_features=len(input_data[0])
number_of_datapoints=len(input_data)
print(input_data)
print(output_data)
```

```python
def cost(x,y,m):
    '''
    m is coefficient array so get cost using input output and coefficient array
    '''
    cost=0
    for i in range(number_of_datapoints):
        cost=cost+(y[i]-(m*x[i]).sum())**2
    return cost/number_of_datapoints
```

```python
def gradient_descent_each_step(x,y,learning_rate,m):
    '''
    Input: x is 2D input array and y is 1D array and m is the coef_ array
    '''
    for j in range(number_of_input_features):
        value_of_m_j=0
        for i in range(number_of_datapoints):
            value_of_m_j+=(y[i]-(m*x[i]).sum())*(x[i][j])
        m[j]=m[j]-((learning_rate)*value_of_m_j*(-2/number_of_datapoints))
    return m
```

```python
def gradient_descent(x,y):
    learning_rate=1
    m=np.zeros(number_of_input_features)
    old_cost=1
    new_cost=0
    iteration_number=0
    while(old_cost-new_cost>=0.000000000000001): #as cost itne decimal tak predict kar rha tha so...
        iteration_number+=1
        old_cost=cost(x,y,m)
        m=gradient_descent_each_step(x,y,learning_rate,m)
        new_cost=cost(x,y,m)
        if(new_cost>old_cost):
            while(new_cost>old_cost):
                old_cost=new_cost
                learning_rate=learning_rate*(0.95)
                m=gradient_descent_each_step(x,y,learning_rate,m)
                new_cost=cost(x,y,m)
        print(cost(x,y,m)," ",iteration_number," ",learning_rate)
    return m
```

```python
m=gradient_descent(input_data,output_data)
print(m)
```

```python
testing_input=np.loadtxt("test.csv",delimiter=",")
testing_input_list=testing_input.tolist()
for i in range(len(testing_input)):
    testing_input_list[i].append(1)
testing_input=np.array(testing_input_list)
print(testing_input.shape)
print(m.shape)
```

```python
print((m*testing_input).shape)
```

```python
test_input=np.asmatrix(testing_input)
coef_=np.transpose(np.asmatrix(m))
```

```python
y_predicted=test_input*coef_
print(y_predicted.shape)
```

```python
output=np.array(y_predicted)
```

```python
y_pred=[]
for i in output:
    y_pred.append(i[0])
print(y_pred)
```
