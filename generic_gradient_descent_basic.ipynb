```python
from sklearn import datasets
import numpy as np
```

```python
diabetes=datasets.load_diabetes()
Input=np.array(diabetes["data"])
Output=np.array(diabetes["target"])
print(Input)
print(Output)
print(Input.shape,Output.shape)
```

```python
def cost(x,y,m):
    M=len(x)
    answer=0
    for i in range(M):
        answer+=(y[i]-(m*x[i]).sum())**2
    return answer/M
```

```python
Input=Input.tolist()
for i in range(len(Input)):
    Input[i].append(1)
Input=np.array(Input)
print(Input.shape)
```

```python
#so input is updated will last mei all 1
def gradient_descent_each_step(x,y,learning_rate,m):
    '''
    Input: x is 2D input array and y is 1D array and m is the coef_ array
    '''
    number_of_data_points=len(x)
    number_of_features=len(x[0])
    for j in range(number_of_features):
        value_of_m_j=0
        for i in range(number_of_data_points):
            value_of_m_j+=(y[i]-(m*x[i]).sum())*(x[i][j])
        m[j]=m[j]-(learning_rate)*value_of_m_j
    return m
```

```python
def gradient_descent(x,y,learning_rate,number_of_iterations):
    m=np.zeros(len(x[0]))
    for i in range(number_of_iterations):
        old_cost=cost(x,y,m)
        m=gradient_descent_each_step(x,y,learning_rate,m)
        new_cost=cost(x,y,m)
        print(cost(x,y,m)," ",i)
    return m
```
