```python
import numpy as np
```

```python
data=np.loadtxt("data.csv",delimiter=",")
Input=data[:,0]
Output=data[:,1]
```

```python
def cost(x,y,m,c):
    '''
    y is output array and x is input array
    '''
    N=len(x)
    return (((y-m*x-c)**2).sum())/N
```

```python
def gradient_descent_each_step(x,y,m,c,learning_rate):
    '''
    x is input array and y is output array
    returns updated m and c after one time m'=m-(alpha)(slope_wrt_m)
    '''
    N=len(x) #number of datapoints
    m=m-(learning_rate)*(-2/N)*(((y-m*x-c)*x).sum())
    c=c-(learning_rate)*(-2/N)*((y-m*x-c).sum())
    return (m,c)
```

```python
def gradient_descent(x,y,learning_rate,number_of_iterations):
    '''
    number of iteration times kar process of updation of input and output
    '''
    #initial random value of m and c taken as 0 each
    m=0
    c=0
    for i in range(number_of_iterations):
        m,c=gradient_descent_each_step(x,y,m,c,learning_rate)
        print("Cost : ",cost(x,y,m,c)," for m=",m," and c=",c)
        #so yeh updated m aur c each iterations
    return m,c
```

```python
m,c=gradient_descent(Input,Output,0.001,100)
print(m)
print(c)
```

```python
#clearly cost increases for every iteration so overshooting ho rhi change alpha
m,c=gradient_descent(Input,Output,0.0001,100)
print(m)
print(c)
```

```python
#now undershoot nahi ho rha but maybe more iterations required to get to ans
#so alpha=0.0005 kar de
m,c=gradient_descent(Input,Output,0.0005,100)
print(m)
print(c)
```

```python
#isse bhi overshoot so 0.0001 to 0.0005
#alpha=0.00025 maybe
m,c=gradient_descent(Input,Output,0.00038,1000)
print(m)
print(c)
```
